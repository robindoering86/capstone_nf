```python
# Randomly selected 50 stocks out of S&P500
symb_list = ['CNC', 'FMC', 'ALGN', 'RTN', 'FAST', 'ETFC', 'ABC', 'PFG', 'AXP', 'PNW',
       'CMS', 'URI', 'PRGO', 'DRI', 'MAS', 'M', 'ZION', 'CINF', 'HIG', 'EOG',
       'ICE', 'CBOE', 'ALL', 'UHS', 'FRT', 'NEM', 'BLL', 'NDAQ', 'CMG', 'JKHY',
       'T', 'QCOM', 'FTI', 'AOS', 'ATVI', 'STT', 'TXN', 'KIM', 'NTAP', 'CHD',
       'WYNN', 'GM', 'SJM', 'MCHP', 'ES', 'KMB', 'GIS', 'EXC', 'IBM', 'MSFT']
```

```python
startd = dt.datetime(2011, 12, 19)
endd = dt.datetime.today()
```

```python
dm = MSDataManager()
```

```python
dm.ReturnData('MSFT', start_date=startd, end_date=dt.datetime.today(), limit=None)
```

```python
start = time.time()
df = pd.DataFrame(pd.MultiIndex.from_product([[], []]))
dic = {} 
#symb_list = dm.ReturnIndexConstituents('DJIA')+['DJIA']
for symb in symb_list:
    #print(symb)
    symb_df = dm.ReturnData(
        symb,
        start_date=startd,
        end_date=endd,
        limit=None)
    dic[symb] = symb_df
    
djia_conc = pd.concat(dic.values(), keys=dic.keys(), axis=1)
print(f'Elapsed time: {int(time.time()-start)}s')
```

```python
for symb in symb_list:
    print(symb, sum(djia_conc[symb].isna().sum()))
```

```python
import pickle
fname = data_base_dir + '/processed/djia_raw_2019-12-11.pkl'
pickle.dump(djia_conc, open(fname, 'wb'))
```
